```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
df = pd.concat([train, test])
```

```python
df.head(2)
```

```python
df.select_dtypes('object').columns
```

```python
df = pd.get_dummies(df, columns=['gender', 'marital_status', 'education_level', 'employment_status','loan_purpose'], dtype='int64', drop_first=True)
```

```python
df.head(2)
```

```python
sns.heatmap(df.isnull())
```

```python
df.isnull().sum()
```

```python
df['loan_paid_back'].fillna(df['loan_paid_back'].mean(), inplace=True)
```

```python
sns.heatmap(df.isnull())
```

```python
df.drop(['grade_subgrade'], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaler.fit(df)
df1 = scaler.transform(df)
```

```python
train_len = train.shape[0]
X_train = df1[:train_len]
X_test = df1[train_len:]
y_train = train['loan_paid_back']
```

```python

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, r2_score
```

```python
def test_model(model, X_train, y_train):
    cv = KFold(n_splits=5, shuffle=True, random_state=42)
    r2 = make_scorer(r2_score)
    r2_val_score = cross_val_score(model, X_train, y_train, cv=cv, scoring=r2)
    score = [r2_val_score.mean()]
    return score
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
test_model(lr, X_train, y_train)
```

```python
from sklearn.linear_model import Ridge, Lasso
ridge = Ridge(alpha=0.1)
test_model(ridge, X_train, y_train)
```

```python
lasso = Lasso(alpha=0.1)
test_model(lasso, X_train, y_train)
```

```python

from sklearn.svm import SVR
svr = SVR()
test_model(svr, X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=42)
test_model(dt, X_train, y_train)
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = np.exp(lr.predict(X_test))
```

```python
y_pred
```

```python
submit_test = pd.concat([test['id'], pd.DataFrame(y_pred, columns=['loan_paid_back'])], axis=1)
submit_test.head()
```

```python
submit_test.to_csv('loan_pay_back_submission.csv', index=False)
```
